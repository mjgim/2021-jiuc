```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams.update({"figure.constrained_layout.use": True})
```

### 선 그래프

주로 (시간 등) 연속으로 수집된 데이터나, 좌표평면 위에서 이동하는 데이터를 그립니다.
마커와 선 양식에 차이를 줄 수 있습니다.

참고: 일반적으로 x축, y축, 그래프 제목이 포함되는 것이 좋습니다.

```python
fig, ax = plt.subplots()

t = np.linspace(0, 4 * np.pi)
ax.plot(t, np.sin(t), label="$\sin(x)$")
ax.plot(t, np.sin(t - 1), "o-.", label="$\sin(x + 1)$")
ax.plot(t, np.sin(t - 2), "s", label="$\sin(x + 1)$")
ax.set(
    xlabel="Time (second)",
    ylabel="Position (mm)",
    title="Position of the ball",
)
ax.legend()
pass
```

### 산점도

주로 데이터 중 두 개의 변수의 관계를 보고 싶을 때 사용합니다.
마커의 크기, 모양, 색깔로 추가 정보를 넣을 수 있습니다.

```python
rng = np.random.default_rng(78)

n = 20
feature_a = rng.normal(size=n)
feature_b = rng.normal(size=n)
feature_nonnegative = rng.uniform(size=n)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.scatter(feature_a, feature_b, s=feature_nonnegative * 1000, alpha=0.1)
ax1.set(
    xlabel="Feature A",
    ylabel="feature B",
    title="Some title (marker size = feature C)"
)

cb = ax2.scatter(feature_a, feature_b, c=feature_nonnegative, cmap="viridis")
ax2cb = fig.colorbar(cb)
ax2.set(
    xlabel="Feature A",
    ylabel="feature B",
    title="Some title"
)
ax2cb.ax.set_ylabel("Feature C")
pass
```

### 막대 그래프

데이터의 값에 비례하는 높이(혹은 길이)를 지닌 막대로 표현하는 그래프입니다.

주의: ‘숫자 세기’를 제외한 거의 대부분의 경우에 다른 더 좋은 그래프가 있습니다.

주의: 거의 대부분의 경우 막대가 시작하는 부분의 축은 0부터 시작해야 합니다.

```python
hamlet = "To be, or not to be, that is the question"
hamlet_count = pd.Series(list(hamlet)).value_counts()

fig, ax = plt.subplots()
ax.bar(hamlet_count.index, hamlet_count)
ax.set(
    xlabel="Character",
    ylabel="Occurrence",
    title=hamlet,
)
pass
```

### 히스토그램

막대 그래프의 특수한 경우로, 도수 분포를 그래프로 나타낸 것입니다.

```python
rng = np.random.default_rng(78)

n = 500
x = np.concatenate((rng.normal(loc=0, size=n), rng.normal(loc=3, size=n)))

fig, axes = plt.subplots(1, 3, figsize=(18, 4))

for bins, ax in zip([5, 40, 200], axes.flat):
    ax.hist(x, bins=bins, density=True, fc="tab:blue", ec="darkblue")
    ax.set(
        xlabel="Value of x",
        ylabel="Probability",
        title=f"Histogram of x, bins = {bins}",
    )
pass
```

### 상자 수염 그림

데이터 분포의 통계량(최대, 최소, 사분위값)을 표현한 그래프입니다.
예를 들어 여러 번 실험한 수치의 통계량을 표현하고 싶을때엔, 막대 그래프보다 상자 그래프가 일반적으로 더 낫습니다.

```python
iris = pd.read_csv(
    "https://gist.githubusercontent.com"
    "/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
)
iris
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 4), sharey=True)

sns.boxplot(data=iris, x="variety", y="sepal.length", ax=ax1)
sns.swarmplot(data=iris, x="variety", y="sepal.length", ax=ax2)
sns.violinplot(data=iris, x="variety", y="sepal.length", ax=ax3)
pass
```

### 히트맵, 등고선

```python
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 90)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) ** 5 + np.cos(X) * np.cos(Y)

fig, ax = plt.subplots()

contours = ax.contour(X, Y, Z, 3, colors="black", levels=5)
ax.clabel(contours, inline=True, fontsize=8)
cb = ax.imshow(
    Z, extent=[x[0], x[-1], y[0], y[-1]], origin="lower", cmap="RdBu", alpha=0.5
)
ax.set(
    xlabel="x",
    ylabel="y",
)
fig.colorbar(cb)
pass
```

## 컬러맵

적절한 색깔과 컬러맵을 사용하면 시각화의 의도를 더 잘 전달할 수 있습니다.
또한 흑백 출력될 가능성에도 주의해야 합니다.

참고 자료:
- [Choosing Colormaps in Matplotlib](https://matplotlib.org/stable/tutorials/colors/colormaps.html)
- [Turbo, An Improved Rainbow Colormap for Visualization ](https://ai.googleblog.com/2019/08/turbo-improved-rainbow-colormap-for.html)
- [Color Blind Vision Simulator](https://pilestone.com/pages/color-blindness-simulator-1)

```python
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 90)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) ** 5 + np.cos(X) * np.cos(Y)

fig, axes = plt.subplots(2, 2, figsize=(8, 8))

for cmap, ax in zip(["rainbow", "jet", "hsv", "twilight"], axes.flat):
    cb = ax.imshow(Z, extent=[x[0], x[-1], y[0], y[-1]], origin="lower", cmap=cmap)
    ax.set(
        xlabel="x",
        ylabel="y",
        title=cmap,
    )
pass
```
