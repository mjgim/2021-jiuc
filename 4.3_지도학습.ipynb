### 결정 트리 학습법

결정 트리 학습법(decision tree learning)은 전체 데이터를 가장 잘 분류할 수 있도록 하는 적절한 분할 기준을 찾아 부분 집합들로 나누는 과정을 반복하는 ‘결정 트리’를 이용해 분류나 회귀 분석을 하는 알고리즘입니다.

### 장점
-  Decision Tree는 분석 과정이 직관적이고 이해하기 쉽다. 대부분의 기계학습은 분석 결과에 대한 설명을 이해하기 어려운 블랙박스 모델인 반면,Decision Tree는 분석과정을 실제로 눈으로 관측할 수 있는 대표적인 화이트박스 모델이다.
- 비교적 다른 알고리즘에 비해 계산비용이 낮아 데규모 데이터에서도 비교적 빠른 연산이 가능하다.
- 수치형 자료와 범주형 자료 모두에 적용이 가능한 모델이다.

### 단점
- 최적의 Tree를 학습시키는 문제는 NPC 문제로 알려져 있다.
<br> $\Longrightarrow$ Greedy 알고리즘 사용: 최적해를 구하는 근사적 방법, 매 순간에 최적이라고 생각되는 것을 선택해 나가는 방식으로 진행
- 많은 변수의 데이터를 일반화하지 않고 사용할 경우 복잡한 결정트리가 만들어질 수 있다.
<br> $\Longrightarrow$ Pruning(가지치기)를 사용하여 Overfitting을 방지 할 수 있다.
- 데이터의 특성이 특정 변수에 수직/수평적으로 구분되지 못할 때 성능이 떨어진다.

# 과대적합 문제

과대적합은 문자 그대로 모델이 너무 과도하게 학습데이터에 학습되어서 새로운 데이터에 맞지 않는(즉, 일반화 되지 않는) 문제를 말한다. 이러한 과대적합문제를 해결하기 위한 방법 중 하나가 Validation을 이용하는 방법이다.

- Figure. Overfitting

## Validation Data_검증데이터

일반적으로 모든 학습데이터로 모델을 학습시키면 시킬 수록 학습데이터에 대한 Error는 줄어든다.

- Figure. Training Error
![](./data/04/04_06.png)
http://rfriend.tistory.com/188

그러나 이 경우 모델이 Underfitting인지(학습이 잘 되었는지), Overfitting인지(학습이 과하게 되었는지) 가늠하기 힘들다.

이 때, 이러한 Overfitting을 탐지하기 위해 사용하는 Data를 Validation Data(검증 데이터)라고 한다.

## How to prevent Overfitting

주어진 Data set을 Training set, Validation set, Test set 3개의 set으로 구분을 한 후

(보통 Train:Validataion:Test의 비율은 5:2.5:2.5 혹은 5:2:2 정도이나 Data 자체가 적을 땐 Training을 더 늘리기도 한다)

 1. Training set을 이용하여 모델을 학습시키고
 2. Validation set을 가지고 1번에서 학습된 모델의 에러를 조사하고, 과적합의 여부를 판단한다.(Validataion의 Error가 감소하다 증가하는 부분에서 Overfitting이 생겼다고 본다)
 3. Test set을 이용하여 1번과 2번을 이용하여 만든 최종 모델의 성능을 파악한다. -> 최종 모델 성능 평가용

# 과대적합 문제_Over-fitting in Decision Tree


일반적으로 Tree는 모든 클래스가 완전히 분리 될 때 까지(모든 Leaf 노드가 순수 노드가 될 때까지) 진행되므로 아래 그림과 같이 모델이 매우 복잡해지고 학습데이터에 과적합(Overfitting)되는 문제가 생긴다.

## Pruning_가지치기

Decision Tree에서 Overfitting 문제를 해결하는 방법은 두가지 방법이 있다.

하나는 학습데이터를 완벽하게 분류하는 시점에 도달하기 전에 Tree를 멈추는 방법(Pre-Pruning)이고 다른 하나는 Tree를 완성한 이후에 가지치기를 하는 방법(Post Pruning)이다.

### Pre-Pruning_사전 가지치기

Pre-Pruning은 Tree의 최대 깊이나 가지의 최대 갯수를 제한하거나, 또는 노드가 분할하기 위한 포인트의 최소 갯수를 지정하는 방법을 말한다.

scikit-learn에서 제공하는 Decision Tree는 Pre-pruning만 지원한다. 그러나 이 방법은 Tree의 성장을 멈춰야 하는 시점을 파악하기 어렵기 때문에 Post Pruning를 많이 사용한다.

### Post-Pruning_사후 가지치기

Post-Pruning에서는 Training set과 Validation set을 사용한다. 여러 종류가 있는데, 대표적인 방법은 다음과 같다.

__Reduced Error Pruning__

각 노드가 모두 pruning의 대상이 되며 해당 노드를 pruning 했을 때의 Tree의 성능이 pruning 하기 전의 Tree보다 성능이 나빠지지 않는 경우 노드의 pruning을 결정한다.(이때 validation set을 사용하여 성능을 측정한다.)

# 코딩

## 1. 라이브러리 불러오기

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import sklearn.datasets as skdatasets
import sklearn.tree as sktree
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor

plt.rcParams.update({"figure.constrained_layout.use": True})
```

## 2. 데이터 불러오기

```python
iris = pd.read_csv('./data/04/iris.csv')
iris_features = iris[["sepal.length", "sepal.width", "petal.length", "petal.width"]]
iris_target = iris["variety"]
iris
```

```python
with plt.rc_context({"figure.constrained_layout.use": False}):
    sns.pairplot(iris, hue="variety")
```

## 3. 학습데이터, 테스트데이터 분류 

```python
X_train, X_test, y_train, y_test = train_test_split(iris_features, iris_target, random_state=78)
```

## 4. 모델 불러오기 및 학습 

```python
clf = DecisionTreeClassifier(random_state=78)
clf.fit(X_train, y_train)
```

```python
print(sktree.export_text(clf, feature_names=list(iris_features.columns), show_weights=True))
```

```python
fig, ax = plt.subplots(figsize=(24, 16))
sktree.plot_tree(
    clf,# 트리모델명
    feature_names=list(iris_features.columns),
    class_names=clf.classes_,
    impurity = True,#gini 계수 출력
    filled=True,# node의 색을 다르게 출력
    ax=ax,
)
pass
```

### Feature importance

```python
feature_imp = clf.feature_importances_
idx = np.arange(len(iris_features.columns))
plt.barh(idx, feature_imp, align='center')
plt.yticks(idx, iris_features.columns)
plt.xlabel('feature importance', size=15)
plt.ylabel('feature', size=15)
plt.show()
```

## 5. 모델 성능평가

### 새로운 데이터 예측

```python
clf.predict([[5.9, 3.0, 5.1, 3.8],[3.9, 3.0, 5.1, 3.1]])
```

```python
iris_predict = clf.predict(X_test)
print('알고리즘 예측 값:{}, 실제 값: {}'.format(iris_predict[0], y_test.values[0]))
```

### 알고리즘 정확도

```python
acc = clf.score(X_test, y_test)
print("Accuracy of model: {}".format(acc))
```

## Pruning

cikit-learn 패키지는 max_depth를 이용한 pre-pruning만 사용 가능하다.(2020.06.14 기준) 

```python
clf_pruning = DecisionTreeClassifier(random_state=78, max_depth = 2)
clf_pruning.fit(X_train, y_train)
```

```python
fig, ax = plt.subplots(figsize=(24, 16))
sktree.plot_tree(
    clf_pruning,
    feature_names=list(iris_features.columns),
    class_names=clf_pruning.classes_,
    filled=True,
    ax=ax,
)
pass
```

```python
acc = clf_pruning.score(X_test, y_test)
print("Accuracy of model: {:.4f}".format(acc))
```
